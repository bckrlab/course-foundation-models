Initial node features:
$$h_{v_1}^{(0)}=(2,1),\;h_{v_2}^{(0)}=(4,0),\;h_{v_3}^{(0)}=(0,3),\;h_{v_4}^{(0)}=(3,1),\;h_{v_5}^{(0)}=(1,0).$$

For this task, your goal is to calculate the embeddings for node $v_1$ over two layers, by hand, step by step, using the GraphSAGE-mean update:
$$h_v^{(l)}=\mathrm{ReLU}\!\Big(W^{(l)}\big[\,h_v^{(l-1)}\;;\;\tfrac{1}{|N(v)|}\textstyle\sum_{u\in N(v)}h_u^{(l-1)}\,\big]\Big).$$

1. Compute the layer-1 embedding $h_{v_1}^{(1)}$ with $$W^{(1)}=\begin{bmatrix}1&0&1&0\\0&1&-2&0\end{bmatrix}.$$


2. You are now given the layer-1 embeddings of $v_1$'s neighbours (provided — do not recompute from layer 0):
    $$h_{v_2}^{(1)}=(4,0),\;h_{v_3}^{(1)}=(2,4),\;h_{v_4}^{(1)}=(1,3),\;h_{v_5}^{(1)}=(1,1).$$
    Compute the layer-2 embedding $h_{v_1}^{(2)}$ with
    $$W^{(2)}=\begin{bmatrix}1&0&1&0\\0&1&1&0\end{bmatrix}.$$


3. Also state which node outside $v_1$'s direct neighbourhood now influences $h_{v_1}^{(2)}$, and why.

## TAG

Let $G = (V, E)$ be the considered undirected graph with $N = \lvert V \rvert$ nodes and adjacency matrix $\mathbf{A}\in\{0, 1\}^{N\times N}$.
Let $\mathbf{X}\in\mathbb{R}^{N\times F}$ be the node-level features and $\mathbf{Y} = \{1, \dots, C\}^N$ the node-level labels with $C$ label classes.
Let $I$ be the feature encoder count and $J$ the structure encoder count.

1. What is the role of the *feature encoders* $\phi^{(i)}$ and of the structural encoders $\psi^{(j)}$? What are their respective inputs? Which feature encoders and structural encoders are used in TAG paper[^TAG]?

[^TAG]: "Bringing Graphs to the Table: Zero-shot Node Classification via Tabular Foundation Models", Hayler et al. 2025, [`arxiv.org/abs/2509.07143`](https://arxiv.org/abs/2509.07143).

2. TAG constructs a table from node features and labels, similar to G2T-FM. How many rows and columns does the table have? Define the input and output dimensions for each node feature encoder map and structure encoder map.

3. Which technical challenges does TAG face when applying a pre-trained tabular foundation model to the generated table? How are these addressed?

4. Briefly summarize: How does *TAG* differ from *G2T-FM*?
