### Temporal Tasks

We consider the same entity graph as before, but now each sale has a timestamp. Each sale node $s$ has:

- feature: $h_s^{(0)} = \text{price}$
- timestamp: $\tau_s$

We want to predict whether a user will churn at time $t$.

1. For a fixed user $u$ and prediction time $t$, which sale nodes should be included in the neighborhood $\mathcal{N}_t(u)$?

2. Suppose we mistakenly include all sales of a user, what would be the issue?

3. Given our task, we believe that recent purchases should matter more than older ones. The current aggregation rule is:

    $$h_u^{(1)} = W \cdot \sum_{s \in \mathcal{N}_t(u)} v_s$$

    which ignores timestamps.

    Propose a modified aggregation rule that takes timestamps into account and assigns higher importance to more recent sales. Briefly explain how your aggregation captures recency.

## Griffin

Griffin turns the RDB-as-graph view into a foundation model with shared, pretrained pieces: a unified encoder, an MPNN, and a unified decoder.

1. Earlier RDB models learn a separate embedding per categorical column. Griffin instead encodes every cell with pretrained encoders shared across all columns and databases. Why is this the change that makes Griffin a foundation model?

2. Griffin feeds each cell's value and its table + column name into the encoder. For the cell 100 in Sales.price: what is the input, and why is the column name needed?

3. The encoder produces one vector per row. If we sent those straight to the decoder, what relational signal would we lose, and what does the MPNN add?
